```python
bound = 5
p = figure(
    title='Click to leave a dot',
    tools='tap',
    x_range=(-bound, bound),
    y_range=(-bound, bound)
)

circle_source = ColumnDataSource(
    data=dict(
        x=[],
        y=[]
    )
)   
p.circle(
    x='x',
    y='y',
    source=circle_source,
    color='red'
) 

line_source = ColumnDataSource(
    data=dict(
        x=[],
        y=[]
    )
)
p.line(
    x='x',
    y='y',
    source=line_source,
    color='black'
)


# Add circle on click
callback = CustomJS(
    args=dict(
        circle_source=circle_source,
        line_source=line_source
    ),
    code="""
    // the event that triggered the callback is cb_obj:
    // The event type determines the relevant attributes
    console.log('Tap event occurred at x-position: ' + cb_obj.x)
    console.log('Tap event occurred at y-position: ' + cb_obj.y)
    const data_circle = circle_source.data;
    const data_line = line_source.data;
    
    // Get location user clicked
    const x = cb_obj.x;
    const y = cb_obj.y;
    
    // Add circle
    data_circle['x'].push(x);
    data_circle['y'].push(y);
    
    // Add line from origin (0, 0) to (x, y)
    data_line['x'].push(0);
    data_line['x'].push(x);
    data_line['y'].push(0);
    data_line['y'].push(y);
    
    // Update plot
    circle_source.change.emit();
    line_source.change.emit();
"""
)


p.js_on_event(Tap, callback)

show(p)
```
